```python
from cil.utilities import dataexample, noise
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import zarr
import dask.array as da
import pandas as pd

from matplotlib.lines import Line2D

# Get colors / markers / functions
colors = ("b", "g", "r", "c", "m", "y", "k")
markers = []
for m in Line2D.markers:
    try:
        if len(m) == 1 and m != " ":
            markers.append(m)
    except TypeError:
        pass

plt.rcParams['lines.linewidth'] = 5
plt.rcParams['lines.markersize'] = 20
plt.rcParams['font.size'] = 30
plt.rcParams['axes.labelsize'] = "x-large"
```

### Load data

```python
data = zarr.load("../data/denoising/noisy_shapes_seed_10_var_0.05.zarr")["gtruth"]
noisy = zarr.load("../data/denoising/noisy_shapes_seed_10_var_0.05.zarr")["noisy"]

alpha = 0.5
tv_denoising_dual = zarr.load("results/tv_denoising_dual_shapes_alpha_{}.zarr".format(alpha))
dual_objective = tv_denoising_dual["dual_objective"]
dual_solution = tv_denoising_dual["denoised"]
```

### Load data for ISTA, ProxSkip

```python
tmp_ista_time = []
probs = [0.01, 0.1, 0.3, 0.5]
tmp_proxskip_time = {i:[] for i in probs}

num_runs = 1

for run in range(num_runs):

    res_ista = zarr.load("results/denoising_runs/denoising_runs_alpha_0.5/ista_run_{}_its_50000.zarr".format(run))    
    tmp_ista_time.append(res_ista["timing"])  

    for prob in probs:
        res_proxskip = zarr.load("results/denoising_runs/denoising_runs_alpha_0.5/proxskip_run_{}_prob_{}_its_50000.zarr".format(run, prob))
        tmp_proxskip_time[prob].append(res_proxskip["timing"])   
```

```python
ista_time_per_run = [np.sum(tmp_ista_time[i]) for i in range(num_runs)]
proxskip_time_per_run = {i:np.sum(tmp_proxskip_time[i], axis=1) for i in probs}
```

```python
average_time_ista = np.mean(np.array(tmp_ista_time), axis=0)
average_time_proxskip = {i: np.mean(tmp_proxskip_time[i], axis=0) for i in probs}

mean_ista_time = np.mean(ista_time_per_run)
mean_proxskip_time = {i: np.mean(proxskip_time_per_run[i]) for i in probs}
```

```python

colors = ['steelblue', 'orange', 'brown', 'green']
num_its = 50000
markevery = 2000
sol_error = lambda x: x/np.sqrt(np.sum(dual_solution**2)) 
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.semilogy(sol_error(res_ista["rse_primal_cvxpy"][:num_its]), label="ISTA", marker=markers[2], markevery=5000, color="red")

for i in range(len(probs)):
    dd = zarr.load("results/denoising_runs/denoising_runs_alpha_0.5/proxskip_run_{}_prob_{}_its_50000.zarr".format(run, probs[i]))
    plt.semilogy(sol_error(dd["rse_primal_cvxpy"][:num_its]), label="ProxSkip (p = {})".format(probs[i]), markevery=4900, color=colors[i], marker = markers[i+2])
    
ax.legend(ncols=2) #bbox_to_anchor=(0.92, 0.97)
ax.grid(which="major")
plt.ylabel(r"$\frac{\|u_{k} - u^{*}\|_{2}}{\|u^{*}\|_{2}}$")
plt.xlabel("Iterations")
# plt.savefig("figures/denoising/ista_vs_proxskip_multiple_p_alpha_{}_iterates_xaxis_iterations.png".format(alpha), bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.semilogy(np.cumsum(average_time_ista), sol_error(res_ista["rse_primal_cvxpy"][:num_its]), label="ISTA", marker=markers[2], markevery=5000, color="red")
for i in range(len(probs)):
    dd = zarr.load("results/denoising_runs/denoising_runs_alpha_0.5/proxskip_run_{}_prob_{}_its_50000.zarr".format(run, probs[i]))
    plt.semilogy(np.cumsum(average_time_proxskip[probs[i]]), sol_error(dd["rse_primal_cvxpy"][:num_its]), label="ProxSkip (p = {})".format(probs[i]), markevery=4900, color=colors[i], marker = markers[i+2])
ax.legend(ncols=2) 
ax.grid(which="major")
plt.ylabel(r"$\frac{\|u_{k} - u^{*}\|_{2}}{\|u^{*}\|_{2}}$")
plt.xlabel("CPU Time (sec)")
# plt.savefig("figures/denoising/ista_vs_proxskip_multiple_p_alpha_{}_iterates_xaxis_time.png".format(alpha), bbox_inches='tight')
plt.show()
```
