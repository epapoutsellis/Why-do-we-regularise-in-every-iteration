```python
tmp_fista_time = []
tmp_ista_time = []
tmp_proxskip_time = []
alpha = 0.55
epsilon = 0.01
prob = 0.04767
num_iterations = 5000
num_runs = 30
tmp = []
for run in range(num_runs):
    
    res_fista = zarr.load("results_denoising/denoising_runs_alpha_{}_epsilon_{}/fista_run_{}_its_{}.zarr".format(alpha, epsilon, run, num_iterations))
    tmp_fista_time.append(res_fista["timing"])

    res_ista = zarr.load("results_denoising/denoising_runs_alpha_{}_epsilon_{}/ista_run_{}_its_{}.zarr".format(alpha, epsilon, run, num_iterations))
    tmp_ista_time.append(res_ista["timing"])    

    res_proxskip = zarr.load("results_denoising/denoising_runs_alpha_{}_epsilon_{}/proxskip_run_{}_prob_{}_its_{}.zarr".format(alpha, epsilon, run, prob, num_iterations))
    tmp_proxskip_time.append(res_proxskip["timing"])  


fista_time_per_run = [np.sum(tmp_fista_time[i]) for i in range(num_runs)]
ista_time_per_run = [np.sum(tmp_ista_time[i]) for i in range(num_runs)]
proxskip_time_per_run = [np.sum(tmp_proxskip_time[i]) for i in range(num_runs)]

mean_fista_time = np.mean(fista_time_per_run)
mean_ista_time = np.mean(ista_time_per_run)
mean_proxskip_time = np.mean(proxskip_time_per_run)
```

```python
res_proxskip["use_prox"]
```

```python
markevery = 200
```

```python
nrse = lambda x : [i/np.sqrt(np.sum(tv_denoising_dual_strongly["denoised"]**2)) for i in x]
nres_fista = np.array(nrse(res_fista["rse_primal_cvxpy"]))
nres_ista = np.array(nrse(res_ista["rse_primal_cvxpy"]))
nres_proxskip = np.array(nrse(res_proxskip["rse_primal_cvxpy"]))
```

```python

fig, ax = plt.subplots(figsize=(20, 10))
ax.semilogy(nres_fista, label="FISTA", color="black", marker = markers[1], markevery=markevery)
ax.semilogy(nres_ista, label="ISTA", color="red", marker = markers[2], markevery=markevery)
ax.semilogy(nres_proxskip, label="ProxSkip (p = {})".format(prob), color="orange", marker = markers[3], markevery=markevery, alpha=0.8)
ax.legend()
ax.grid(which="major")
# ax.grid(which = "minor")
plt.ylabel(r"$\frac{\|u_{k} - u^{*}\|_{2}}{\|u^{*}\|_{2}}$")
plt.xlabel("Iterations")
plt.savefig("figures/fista_ista_proxskip_p_{}_strongly_alpha_{}_epsilon_{}_iterates_xaxis_iterations.png".format(prob, alpha, epsilon), bbox_inches='tight')
plt.show()

```

```python
# mean and var
average_time_fista = da.mean(np.array(tmp_fista_time), axis=0).compute()
average_time_ista = da.mean(np.array(tmp_ista_time), axis=0).compute()
average_time_proxskip = da.mean(np.array(tmp_proxskip_time), axis=0).compute()


thres = np.where(nres_fista<=1e-6)[0][0]
fista_time_below_thres = np.sum(average_time_fista[:thres])

thres = np.where(nres_ista<=1e-6)[0][0]
ista_time_below_thres = np.sum(average_time_ista[:thres])

thres = np.where(nres_proxskip<=1e-6)[0][0]
proxskip_time_below_thres = np.sum(average_time_proxskip[:thres])

print(fista_time_below_thres, ista_time_below_thres, proxskip_time_below_thres)
```

```python

fig, ax = plt.subplots(figsize=(20,10))
ax.semilogy(np.cumsum(average_time_fista), nrse(res_fista["rse_primal_cvxpy"][:-1]), label="FISTA", color="black", marker = markers[2], markevery=markevery)
ax.semilogy(np.cumsum(average_time_ista), nrse(res_ista["rse_primal_cvxpy"][:-1]), label="ISTA", color="red", marker = markers[2], markevery=markevery)
ax.semilogy(np.cumsum(average_time_proxskip), nrse(res_proxskip["rse_primal_cvxpy"][:-1]), label="ProxSkip (p = {})".format(prob), color="orange", marker = markers[3], markevery=markevery)
ax.axvline(x=fista_time_below_thres, color="black", linestyle="--")
ax.axvline(x=ista_time_below_thres, color="red", linestyle="--")
ax.axvline(x=proxskip_time_below_thres, color="orange", linestyle="--")

ax.text(fista_time_below_thres-3.45, 0.45, r"$\overline{{T}} = {:.3f}$".format(fista_time_below_thres), color='black', ha='center', va='bottom', rotation=90, transform=ax.transAxes)
ax.text(ista_time_below_thres-5.89, 0.45, r"$\overline{{T}} = {:.3f}$".format(ista_time_below_thres), color='red', ha='center', va='bottom', rotation=90, transform=ax.transAxes)
ax.text(proxskip_time_below_thres-3.88,0.45, r"$\overline{{T}} = {:.3f}$".format(proxskip_time_below_thres), color='orange', ha='center', va='bottom', rotation=90, transform=ax.transAxes)
# ax.legend(bbox_to_anchor=(0.067, 0.99), ncols=3)
ax.grid(which="major")
# # ax.grid(which = "minor")
plt.ylabel(r"$\frac{\|u_{k} - u^{*}\|_{2}}{\|u^{*}\|_{2}}$")
plt.xlabel("CPU Time (sec)")
plt.savefig("figures/fista_ista_proxskip_p_{}_strongly_alpha_{}_epsilon_{}_iterates_xaxis_time.png".format(prob, alpha, epsilon), bbox_inches='tight')
plt.show()

```

```python
bar_width = 0.3

fig, ax = plt.subplots(figsize=(30, 10))
ax.bar(np.arange(num_runs) - bar_width, fista_time_per_run, width=bar_width, label="AProjGD", color="black")
ax.bar(np.arange(num_runs), ista_time_per_run, width=bar_width, label="ProjGD", color="red")
ax.bar(np.arange(num_runs) + bar_width, proxskip_time_per_run, width=bar_width, label="ProxSkip (p = {})".format(prob), color="orange")
ax.axhline(y=mean_fista_time, xmin=0.016, xmax=0.954, color="black", linestyle="--")
ax.axhline(y=mean_ista_time, xmin=0.016, xmax=0.954, color="red", linestyle="--")
ax.axhline(y=mean_proxskip_time, xmin=0.016, xmax=0.954, color="orange", linestyle="--")
plt.text(32, mean_fista_time + 0.30, r"$\overline{{T}} = {:.2f}$".format(mean_fista_time), color='black', ha='center', va='bottom')
plt.text(32, mean_ista_time - 0.75, r"$\overline{{T}} = {:.2f}$".format(mean_ista_time), color='red', ha='center', va='bottom')
plt.text(32, mean_proxskip_time -  0.75, r"$\overline{{T}} = {:.2f}$".format(mean_proxskip_time), color='orange', ha='center', va='bottom')
ax.grid("on")
ax.set_xlabel('Runs')
ax.set_ylabel('CPU Time (sec)')
# ax.set_ylim(0, 10)
ax.set_xlim(-1, 34)
ax.set_xticks(np.arange(num_runs))  
ax.set_xticklabels(np.arange(num_runs), rotation=70)  
ax.legend(bbox_to_anchor=(0.2, 0.99), ncols=3)
plt.savefig("figures/fista_ista_proxskip_p_{}_strongly_alpha_{}_epsilon_{}_bar_time.png".format(prob, alpha, epsilon), bbox_inches='tight')
plt.show()
```
