```python
from cil.utilities import dataexample, noise
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import zarr
import dask.array as da


from matplotlib.lines import Line2D

# Get colors / markers / functions
colors = ("b", "g", "r", "c", "m", "y", "k")
markers = []
for m in Line2D.markers:
    try:
        if len(m) == 1 and m != " ":
            markers.append(m)
    except TypeError:
        pass

plt.rcParams['lines.linewidth'] = 5
plt.rcParams['lines.markersize'] = 20
plt.rcParams['font.size'] = 30
```

### Load and show data

```python
data = zarr.load("../data/denoising/noisy_shapes_seed_10_var_0.05.zarr")["gtruth"]
noisy = zarr.load("../data/denoising/noisy_shapes_seed_10_var_0.05.zarr")["noisy"]
```

```python
alpha = 0.5
tv_denoising_dual = zarr.load("results/tv_denoising_dual_shapes_alpha_{}.zarr".format(alpha))
dual_objective = tv_denoising_dual["dual_objective"]
dual_solution = tv_denoising_dual["denoised"]
```

```python
plt.figure()
plt.imshow(tv_denoising_dual["denoised"], cmap="gray")
plt.axis("off")
# plt.savefig("figures/tv_denoised_dual_shapes_alpha_0.5.png", bbox_inches="tight")
plt.show()
```

```python
sol_error = lambda x: x/np.sqrt(np.sum(dual_solution**2))
```

```python
tmp_fista_time = []
tmp_ista_time = []
tmp_proxskip_time = []
prob = 0.1
num_runs = 30

for run in range(num_runs):
    
    res_fista = zarr.load("results/denoising_runs_alpha_0.5/fista_run_{}_its_50000.zarr".format(run))
    tmp_fista_time.append(res_fista["timing"])

    res_ista = zarr.load("results/denoising_runs_alpha_0.5/ista_run_{}_its_50000.zarr".format(run))
    tmp_ista_time.append(res_ista["timing"])    

    # only for p = 0.1
    res_proxskip = zarr.load("results/denoising_runs_alpha_0.5/proxskip_run_{}_prob_{}_its_50000.zarr".format(run, prob))
    tmp_proxskip_time.append(res_proxskip["timing"])    


# fista_time_per_run = [np.sum(tmp_fista_time[i]) for i in range(num_runs)]
ista_time_per_run = [np.sum(tmp_ista_time[i]) for i in range(num_runs)]
proxskip_time_per_run = [np.sum(tmp_proxskip_time[i]) for i in range(num_runs)]

# mean_fista_time = np.mean(fista_time_per_run)
mean_ista_time = np.mean(ista_time_per_run)
mean_proxskip_time = np.mean(proxskip_time_per_run)
```

```python
markevery = 2000
```

```python

fig, ax = plt.subplots(figsize=(20, 10))
# ax.semilogy(res_fista["rse_primal_cvxpy"][:-1], label="AProjGD", color="black", marker = markers[1], markevery=markevery)
ax.semilogy(sol_error(res_ista["rse_primal_cvxpy"]), label="ProjGD", color="red", marker = markers[2], markevery=markevery)
ax.semilogy(sol_error(res_proxskip["rse_primal_cvxpy"]), label="ProxSkip (p = 0.1)", color="orange", marker = markers[3], markevery=markevery, alpha=0.8)
ax.legend()
ax.grid(which="major")
# ax.grid(which = "minor")
plt.ylabel(r"$\frac{\|u_{k} - u^{*}\|_{2}}{\|u^{*}\|_{2}}$")
plt.xlabel("Iterations")
# plt.savefig("figures/ista_vs_proxskip_p_0.1_alpha_{}_iterates_xaxis_iterations.png".format(alpha), bbox_inches='tight')
plt.show()

```

```python
seed = 40
rng = np.random.default_rng(seed)
theta = rng.choice([1,0], p=[prob,1-prob], size=50000)

vert1 = np.array(theta[0:101]).astype("int32")
ind1 = np.where((vert1 == 1))[0][1:]

vert2 = np.array(theta[5000:5101]).astype("int32")
ind2 = np.where((vert2 == 1))[0]

vert3 = np.array(theta[49900:]).astype("int32")
ind3 = np.where((vert3 == 1))[0]


fig, ax = plt.subplots(figsize=(20, 10))
ax.semilogy(sol_error(res_ista["rse_primal_cvxpy"][0:101]), label="ISTA", color="red", marker = markers[2], markevery=ind1)
ax.semilogy(sol_error(res_proxskip["rse_primal_cvxpy"][0:101]), label="ProxSkip (p = 0.1)", color="orange", marker = markers[3], markevery=ind1)
ax.set_ylabel(r"$\frac{\|u_{k} - u^{*}\|_{2}}{\|u^{*}\|_{2}}$", labelpad=20)
ax.set_xlabel("Iterations")
ax.legend(loc="upper right")
for x in ind1:
    ax.axvline(x, color='black', linestyle='--', alpha=0.7, linewidth=2)
ax.grid(which="major")
ax.grid(which = "minor")
ax.set_xticks(np.arange(0,101,5)) 
ax.set_xticklabels(np.arange(0,101,5), rotation=70)
ax.tick_params(axis='x', direction='in', length=5) 
# plt.savefig("figures/ista_vs_proxskip_p_0.1_alpha_{}_iterates_xaxis_iterations_zoom1.png".format(alpha), bbox_inches='tight')
plt.show()


fig, ax = plt.subplots(figsize=(20, 10))
ax.semilogy(sol_error(res_ista["rse_primal_cvxpy"][5000:5101]), label="ISTA", color="red", marker = markers[2], markevery=ind2)
ax.semilogy(sol_error(res_proxskip["rse_primal_cvxpy"][5000:5101]), label="ProxSkip (p = 0.1)", color="orange", marker = markers[3], markevery=ind2)
ax.set_ylabel(r"$\frac{\|u_{k} - u^{*}\|_{2}}{\|u^{*}\|_{2}}$", labelpad=20)
ax.set_xlabel("Iterations")
for x in ind2:
    ax.axvline(x, color='black', linestyle='--', alpha=0.7, linewidth=2)
ax.grid(which="major")
ax.grid(which = "minor")
ax.set_xticks(np.arange(0,101,5))
ax.set_xticklabels(np.arange(5000,5101,5), rotation=70)
# plt.savefig("figures/ista_vs_proxskip_p_0.1_alpha_{}_iterates_xaxis_iterations_zoom2.png".format(alpha), bbox_inches='tight')
plt.show()

fig, ax = plt.subplots(figsize=(20, 10))
ax.semilogy(sol_error(res_ista["rse_primal_cvxpy"][49900:]), label="ISTA", color="red", marker = markers[2], markevery=ind3)
ax.semilogy(sol_error(res_proxskip["rse_primal_cvxpy"][49900:]), label="ProxSkip (p = 0.1)", color="orange", marker = markers[3], markevery=ind3)
ax.set_ylabel(r"$\frac{\|u_{k} - u^{*}\|_{2}}{\|u^{*}\|_{2}}$", labelpad=20)
ax.set_xlabel("Iterations")
for x in ind3:
    ax.axvline(x, color='black', linestyle='--', alpha=0.7, linewidth=2)
ax.grid(which="major")
ax.grid(which = "minor")
ax.set_xticks(np.arange(0,101,5))
ax.set_xticklabels(np.arange(49900,50001,5), rotation=70)
# plt.savefig("figures/ista_vs_proxskip_p_0.1_alpha_{}_iterates_xaxis_iterations_zoom3.png".format(alpha), bbox_inches='tight')
plt.show()


```

```python

bar_width = 0.4

fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(np.arange(num_runs) - bar_width/2, ista_time_per_run, width=bar_width, label="ProjGD", color="red")
ax.bar(np.arange(num_runs) + bar_width/2, proxskip_time_per_run, width=bar_width, label="ProxSkip (p = 0.1)", color="orange")
ax.axhline(y=mean_ista_time, xmin=0.045, xmax=0.954, color="black", linestyle="--")
ax.axhline(y=mean_proxskip_time, xmin=0.045, xmax=0.954, color="black", linestyle="--")
plt.text(5, mean_ista_time + 2.5, r"$\overline{{T}} = {:.2f}$".format(mean_ista_time), color='black', ha='center', va='bottom')
plt.text(5, mean_proxskip_time - 9.5, r"$\overline{{T}} = {:.2f}$".format(mean_proxskip_time), color='black', ha='center', va='bottom')
ax.grid("on")
ax.set_xlabel('Runs')
ax.set_ylabel('CPU Time (sec)')
ax.set_ylim(0, 100)
ax.set_xticks(np.arange(num_runs))  
ax.set_xticklabels(np.arange(num_runs), rotation=70)  
ax.legend()
# plt.savefig("figures/ista_vs_proxskip_p_0.1_alpha_{}_bar_time.png".format(alpha), bbox_inches='tight')
plt.show()
```
