```python
from cil.utilities import dataexample, noise
import numpy as np
import matplotlib.pyplot as plt
import zarr
```

```python
# Load data
data = dataexample.SHAPES.get()

var = 0.05
seed = 10

# Add gaussian noise
noisy_data = noise.gaussian(data, seed = seed, var = var)     
```

```python
plt.imshow(noisy_data.array)
```

```python
dd = zarr.open_group("../data/denoising/noisy_shapes_seed_{}_var_{}.zarr".format(seed, var))
dd["gtruth"] = data.array
dd["noisy"] = noisy_data.array
```
