```python
import cvxpy
from utils_cvxpy import tv_denoising_dual, tv_denoising_primal, tv_denoising_dual_strong_convex
import numpy as np
import zarr
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as PSNR
from skimage.metrics import structural_similarity as SSIM
```

### Load data

```python
noisy_data = zarr.load("../data/denoising/noisy_shapes_seed_10_var_0.05.zarr")["noisy"]

# alpha2 = 0.65
# epsilon = 0.025

alpha2 = 0.55
epsilon = 0.01
```

### Solve dual-TV and dual-Huber TV with CVXpy

```python
denoised_image_dual, _, _, obj_dual = tv_denoising_dual(noisy_data, alpha2, verbose=False)
denoised_image_dual_strongly_convex, q1, q2, obj_dual_strongly_convex = tv_denoising_dual_strong_convex(noisy_data, alpha2, epsilon, verbose=False)
```

### Save to zarr

```python
dd = zarr.open_group("results/tv_denoising_dual_strongly_convex_shapes_alpha_{}.zarr".format(alpha2))
dd["solution"] = [q1, q2]
dd["denoised"] = denoised_image_dual_strongly_convex
dd["dual_objective"] = obj_dual_strongly_convex
```

```python
plt.figure()
plt.imshow(denoised_image_dual)
plt.title("Dual solution")
plt.colorbar()

plt.figure()
plt.imshow(denoised_image_dual_strongly_convex)
plt.title("Dual solution (Strongly-HuberTV)")
plt.colorbar()
```

```python
a1 = SSIM(zarr.load("../data/denoising/noisy_shapes_seed_10_var_0.05.zarr")["gtruth"], denoised_image_dual,  data_range=1.0)
print(f"SSIM (dual-TV) = {a1}")
```

```python
a2 = SSIM(zarr.load("../data/denoising/noisy_shapes_seed_10_var_0.05.zarr")["gtruth"], denoised_image_dual_strongly_convex, data_range=1.0)
print(f"SSIM (dual Huber-TV) = {a2}")
```
